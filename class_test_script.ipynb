```python
import classification_final
import pandas as pd
import copy

data = pd.read_csv('hotel-tas-test-classification.csv')
#nlp_data = copy.deepcopy(data)
```

```python
X = data.drop('Reviewer_Score', axis=1)
Y = data['Reviewer_Score']
cc_object = classification_final.preprocess(X, Y, data)
result = cc_object.pipeline()
feat = classification_final.featureEningeering(X, X)
feat.engineering_test()

a=classification_final.load_model('classification_models/rndclass.sav') .predict(feat.xtrain)
b=classification_final.load_model('classification_models/logclass.sav') .predict(feat.xtrain)
c=classification_final.load_model('classification_models/treeclass.sav').predict(feat.xtrain)
d=classification_final.load_model('classification_models/gbclass.sav')  .predict(feat.xtrain)
e=classification_final.load_model('classification_models/svmclass.sav') .predict(feat.xtrain)


```

```python
from sklearn.metrics import accuracy_score


ground_truth_labels = cc_object.Y  # Assuming Y contains the ground truth labels

print("Random Forest Accuracy:", accuracy_score(ground_truth_labels, a))
print("Logistic Regression Accuracy:", accuracy_score(ground_truth_labels, b))
print("Decision Tree Accuracy:", accuracy_score(ground_truth_labels, c))
print("Gradient Boosting Accuracy:", accuracy_score(ground_truth_labels, d))
print("Support Vector Machine Accuracy:", accuracy_score(ground_truth_labels,e))
```
